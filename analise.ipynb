```python
#juntando os arquivos excluindo a primeira linha
def concat(arquivo_saida, *arquivos_entrada):
    with open(arquivo_saida, 'w') as saida:
        for arquivo_entrada in arquivos_entrada:
            with open(arquivo_entrada, 'r') as entrada:
                next(entrada)
                for linha in entrada:
                    saida.write(linha)
                    
concat('cotacao.txt', 'COTAHIST_A2023.TXT', 'COTAHIST_A2024.TXT')
```
